```python
from fake_useragent import UserAgent
from lxml.html import fromstring
import requests
from itertools import cycle
import traceback

def get_proxies():
    url = 'https://free-proxy-list.net/'
    response = requests.get(url)
    parser = fromstring(response.text)
    proxies = set()
    for i in parser.xpath('//tbody/tr')[:10]:
        if i.xpath('.//td[7][contains(text(),"yes")]'):
            proxy = ":".join([i.xpath('.//td[1]/text()')[0], i.xpath('.//td[2]/text()')[0]])
            proxies.add(proxy)
    return proxies

def getpage(url):
    req_headers = {
    'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
    'accept-encoding': 'gzip, deflate, br',
    'accept-language': 'en-US,en;q=0.8',
    'upgrade-insecure-requests': '1',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'
}
    proxies = get_proxies()
    proxy_pool = cycle(proxies)
    for i in range(1, 11):
        proxy = next(proxy_pool)
    with requests.Session() as s:
        url = url
        r = s.get(url, headers=req_headers, proxies={"http": proxy, "http": proxy})
        text = r.text
        return BeautifulSoup(text, features = 'lxml')
```

```python
import time
import random

def getpages(url):
    result = []
    while url != "no more page":
        page = getpage(url)
        result = result + [page]
        url = getnextpage(page)
        time.sleep(.5+2*random.random())
    return result
```

```python
def extractpage(page):
    headers = ['property_type', 'price', 'bed', 'bath', 'sqft', 'address', 'link', 'solddate']
    listings = getlistings(page)
    info_dict_list = []
    for i in range(len(listings)):
        property_type = gettype(listings[i])
        solddate = getsolddate(listings[i])
        price = getprice(listings[i])
        bed = gethouseBed(listings[i])
        bath = gethouseBath(listings[i])
        sqft = getsqft(listings[i])
        address = getaddress(listings[i])
        link = getlink(listings[i])
        info_dict = dict(zip(headers, [property_type, price, bed, bath, sqft, address, link, solddate]))
        info_dict_list.append(info_dict)
    return pd.DataFrame(info_dict_list)
```

```python
def extractpages(linklist):
    pages = []
    for link in linklist:
        pages.append(getpage(link))
    df = extractpage(pages[0])
    for i in range(1, len(pages)):
        df = df.append(extractpage(pages[i]))
    return df
```

**Single listing Info**

```python
def listinginfo(page):
    headers = ['built_year', 'renovated_year', 'lotsize', 'story', 'garage', 'parking']
    built_year = getyearbuilt(page)
    renovated_year = getrenovated(page)
    lotsize = getlotsize(page)
    story = getstories(page)
    garage = getgarage(page)
    parking = getparking(page)
    info_dict = dict(zip(headers, [built_year, renovated_year, lotsize, story, garage, parking]))
    return info_dict
```
